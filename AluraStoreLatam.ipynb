### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

#tienda.head()
tienda2.head()
```

#1. Análisis de facturación



```python
ingreso_total_tienda1 = tienda['Precio'].sum()
ingreso_total_tienda2 = tienda2['Precio'].sum()
ingreso_total_tienda3 = tienda3['Precio'].sum()
ingreso_total_tienda4 = tienda4['Precio'].sum()

print(f"Ingreso total Tienda 1: {ingreso_total_tienda1}")
print(f"Ingreso total Tienda 2: {ingreso_total_tienda2}")
print(f"Ingreso total Tienda 3: {ingreso_total_tienda3}")
print(f"Ingreso total Tienda 4: {ingreso_total_tienda4}")
```

# 2. Ventas por categoría

```python
ventas_por_categoria_tienda1 = tienda.groupby('Categoría del Producto').size().sort_values(ascending=False)
ventas_por_categoria_tienda2 = tienda2.groupby('Categoría del Producto').size().sort_values(ascending=False)
ventas_por_categoria_tienda3 = tienda3.groupby('Categoría del Producto').size().sort_values(ascending=False)
ventas_por_categoria_tienda4 = tienda4.groupby('Categoría del Producto').size().sort_values(ascending=False)

print("\nVentas por Categoría - Tienda 1:")
print(ventas_por_categoria_tienda1)

print("\nVentas por Categoría - Tienda 2:")
print(ventas_por_categoria_tienda2)

print("\nVentas por Categoría - Tienda 3:")
print(ventas_por_categoria_tienda3)

print("\nVentas por Categoría - Tienda 4:")
print(ventas_por_categoria_tienda4)
```

# 3. Calificación promedio de la tienda


```python
calificacion_promedio_tienda1 = tienda['Calificación'].mean()
calificacion_promedio_tienda2 = tienda2['Calificación'].mean()
calificacion_promedio_tienda3 = tienda3['Calificación'].mean()
calificacion_promedio_tienda4 = tienda4['Calificación'].mean()

print(f"Calificación promedio Tienda 1: {calificacion_promedio_tienda1:.2f}")
print(f"Calificación promedio Tienda 2: {calificacion_promedio_tienda2:.2f}")
print(f"Calificación promedio Tienda 3: {calificacion_promedio_tienda3:.2f}")
print(f"Calificación promedio Tienda 4: {calificacion_promedio_tienda4:.2f}")
```

# 4. Productos más y menos vendidos

```python
for i, store_df in enumerate([tienda, tienda2, tienda3, tienda4]):
    store_name = f"Tienda {i+1}"
    print(f"\n--- {store_name} ---")

    # Calculate sales count for each product
    product_sales = store_df['Producto'].value_counts()

    # Top 5 most sold products
    print("Productos más vendidos (Top 5):")
    print(product_sales.head(5))

    # Bottom 5 least sold products
    print("\nProductos menos vendidos (Bottom 5):")
    print(product_sales.tail(5))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

stores = {
    "Tienda 1": tienda,
    "Tienda 2": tienda2,
    "Tienda 3": tienda3,
    "Tienda 4": tienda4
}

plt.figure(figsize=(15, 12))

for i, (store_name, df) in enumerate(stores.items()):
    plt.subplot(2, 2, i + 1)
    top_products = df['Producto'].value_counts().head(5)
    sns.barplot(x=top_products.values, y=top_products.index, palette='viridis')
    plt.title(f'Top 5 Productos Más Vendidos - {store_name}')
    plt.xlabel('Cantidad Vendida')
    plt.ylabel('Producto')
    plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

# 5. Envío promedio por tienda

```python
envio_promedio_tienda1 = tienda['Costo de envío'].mean()
envio_promedio_tienda2 = tienda2['Costo de envío'].mean()
envio_promedio_tienda3 = tienda3['Costo de envío'].mean()
envio_promedio_tienda4 = tienda4['Costo de envío'].mean()

print(f"Costo de envío promedio Tienda 1: {envio_promedio_tienda1:.2f}")
print(f"Costo de envío promedio Tienda 2: {envio_promedio_tienda2:.2f}")
print(f"Costo de envío promedio Tienda 3: {envio_promedio_tienda3:.2f}")
print(f"Costo de envío promedio Tienda 4: {envio_promedio_tienda4:.2f}")
```

## Generando gráficos

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 1. Gráfico de Ingreso Total por Tienda
ingresos = {
    "Tienda 1": ingreso_total_tienda1,
    "Tienda 2": ingreso_total_tienda2,
    "Tienda 3": ingreso_total_tienda3,
    "Tienda 4": ingreso_total_tienda4
}
ingresos_df = pd.DataFrame(list(ingresos.items()), columns=['Tienda', 'Ingreso Total'])

plt.figure(figsize=(10, 6))
sns.barplot(x='Tienda', y='Ingreso Total', data=ingresos_df, palette='rocket')
plt.title('Ingreso Total por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Ingreso Total')
plt.ticklabel_format(style='plain', axis='y') # Evita la notación científica en el eje Y
plt.grid(axis='y', linestyle='--')
plt.show()

# 2. Gráfico de Ventas por Categoría de Producto (comparativo entre tiendas)
categorias_df_list = []
for i, ventas_serie in enumerate([ventas_por_categoria_tienda1, ventas_por_categoria_tienda2, ventas_por_categoria_tienda3, ventas_por_categoria_tienda4]):
    temp_df = ventas_serie.reset_index()
    temp_df.columns = ['Categoria', 'Cantidad Vendida']
    temp_df['Tienda'] = f'Tienda {i+1}'
    categorias_df_list.append(temp_df)

ventas_categorias_total_df = pd.concat(categorias_df_list)

plt.figure(figsize=(14, 8))
sns.barplot(x='Categoria', y='Cantidad Vendida', hue='Tienda', data=ventas_categorias_total_df, palette='viridis')
plt.title('Cantidad de Productos Vendidos por Categoría y Tienda')
plt.xlabel('Categoría del Producto')
plt.ylabel('Cantidad Vendida')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Tienda')
plt.tight_layout()
plt.show()

# 3. Gráfico de Calificación Promedio por Tienda
calificaciones = {
    "Tienda 1": calificacion_promedio_tienda1,
    "Tienda 2": calificacion_promedio_tienda2,
    "Tienda 3": calificacion_promedio_tienda3,
    "Tienda 4": calificacion_promedio_tienda4
}
calificaciones_df = pd.DataFrame(list(calificaciones.items()), columns=['Tienda', 'Calificación Promedio'])

plt.figure(figsize=(10, 6))
sns.barplot(x='Tienda', y='Calificación Promedio', data=calificaciones_df, palette='mako')
plt.title('Calificación Promedio por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Calificación Promedio')
plt.ylim(0, 5) # Las calificaciones suelen estar entre 1 y 5
plt.grid(axis='y', linestyle='--')
plt.show()
```

## Preparar Datos Geográficos Combinados

### Subtask:
Combinar los DataFrames de todas las tiendas (`tienda`, `tienda2`, `tienda3`, `tienda4`) en uno solo, añadiendo una columna que identifique a qué tienda pertenece cada venta, para facilitar el análisis geográfico conjunto.


**Reasoning**:
To prepare for combining the dataframes, I will first add a 'Tienda' column to each individual store's dataframe to identify its origin, as instructed in the subtask.



```python
tienda['Tienda'] = 'Tienda 1'
tienda2['Tienda'] = 'Tienda 2'
tienda3['Tienda'] = 'Tienda 3'
tienda4['Tienda'] = 'Tienda 4'
```

**Reasoning**:
Now that each DataFrame has a 'Tienda' column, I will concatenate them into a single DataFrame called `df_total` to facilitate joint geographical analysis, as instructed by the subtask.



```python
df_total = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)
print("Combined DataFrame created with shape:", df_total.shape)
df_total.head()
```

## Visualizar la Distribución Geográfica de Ventas y Métricas Clave

### Subtask:
Generar gráficos de dispersión o mapas de calor usando las coordenadas de latitud y longitud. Estos gráficos mostrarán la distribución de las ventas y permitirán explorar cómo varían los ingresos (Precio) y las calificaciones de los clientes en diferentes ubicaciones geográficas.


**Reasoning**:
To visualize the geographical distribution of sales, income, and customer ratings, I will generate three scatter plots using the combined dataframe `df_total` as instructed by the subtask.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Gráfico de dispersión de la distribución geográfica de todas las ventas por tienda
plt.figure(figsize=(12, 8))
sns.scatterplot(x='lon', y='lat', hue='Tienda', data=df_total, s=20, alpha=0.6)
plt.title('Distribución Geográfica de Ventas por Tienda')
plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.legend(title='Tienda')
plt.grid(True)
plt.show()

# 2. Gráfico de dispersión para mostrar cómo los ingresos (Precio) varían geográficamente
plt.figure(figsize=(12, 8))
sns.scatterplot(x='lon', y='lat', hue='Precio', size='Precio', data=df_total, palette='viridis', sizes=(20, 400), alpha=0.7)
plt.title('Distribución Geográfica de Ingresos (Precio)')
plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.legend(title='Precio', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()

# 3. Gráfico de dispersión para visualizar cómo las calificaciones de los clientes se distribuyen geográficamente
plt.figure(figsize=(12, 8))
sns.scatterplot(x='lon', y='lat', hue='Calificación', size='Calificación', data=df_total, palette='coolwarm', sizes=(20, 400), alpha=0.7)
plt.title('Distribución Geográfica de Calificaciones de Clientes')
plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.legend(title='Calificación', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```
